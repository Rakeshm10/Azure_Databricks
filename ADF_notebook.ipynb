```python
from pyspark.sql import *
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df = spark.read.format('csv').\
    option("header", "true").\
        option("inferSchema", "true").\
            load('abfss://mycontainer@storagedatabricks1005.dfs.core.windows.net/AD_SRC_TGT/source/customers.csv')
```

```python
display(df)
```

```python
df = df.withColumn('full name', concat(col('first_name'), lit(' '), col('last_name')))
```

```python
display(df)
```

```python
df = df.select("customer_id","first_name","last_name","full name","city","phone_number")
```

```python
df = df.withColumnRenamed("full name","Full_name")
```

```python
display(df)
```

```python
df.write.format('Delta').\
    option("header", "true").\
        option("inferSchema", "true").\
            mode('append').save('abfss://mycontainer@storagedatabricks1005.dfs.core.windows.net/AD_SRC_TGT/target/customer/')
```
