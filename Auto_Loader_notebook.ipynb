####Auto Loader

**Streaming data frame**

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df = spark.readStream.format("cloudFiles")\
    .option("cloudFiles.format", "csv")\
        .option("cloudFiles.schemaLocation", "abfss://deltalake@storagedatabricks1005.dfs.core.windows.net/target/check")\
            .load("abfss://deltalake@storagedatabricks1005.dfs.core.windows.net/source")
```

```python
df.writeStream.format("parquet")\
    .option("checkpointLocation", "abfss://deltalake@storagedatabricks1005.dfs.core.windows.net/target/check")\
        .trigger(processingTime="10 seconds")\
        .start("abfss://deltalake@storagedatabricks1005.dfs.core.windows.net/target/data")
```
