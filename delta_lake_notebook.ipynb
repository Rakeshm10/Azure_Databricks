####Delta Table

```python
%sql
create table man_catlog.man_schema.delta_table(id int, name varchar(20),city varchar(50)) using delta
location 'abfss://deltalake@storagedatabricks1005.dfs.core.windows.net/delta_table'
```

**Turn Off Deletion vector**

```python
%sql
ALTER TABLE man_catlog.man_schema.delta_table SET TBLPROPERTIES ('delta.enableDeletionVectors' = false);
```

**Insert data in table**

```python
%sql
insert into man_catlog.man_schema.delta_table values(1,'Ravi','Bangalore'),(2,'Rajesh','Mumbai'),(3,'Rakesh','Delhi'),(4,'Rajesh','Mumbai'),(5,'Ravi','Bangalore'),(6,'Rajesh','Mumbai'),(7,'Rakesh','Delhi'),(8,'Rajesh','Mumbai'),(9,'Ravi','Bangalore');
```

```python
%sql
describe extended man_catlog.man_schema.delta_table
```

```python
%sql
select * from man_catlog.man_schema.delta_table
```

```python
%sql
update man_catlog.man_schema.delta_table set city = 'hyderabad'  where id = 4
```

```python
%sql
select * from man_catlog.man_schema.delta_table
```

**Versioning of table**

```python
%sql
describe history man_catlog.man_schema.delta_table
```

**Turn On Deletion Vector**

```python
%sql
create table man_catlog.man_schema.delta_table2(id int, name varchar(20),city varchar(50)) using delta
location 'abfss://deltalake@storagedatabricks1005.dfs.core.windows.net/delta_table2'
```

```python
%sql
insert into man_catlog.man_schema.delta_table2 values(1,'Ravi','Bangalore'),(2,'Rajesh','Mumbai'),(3,'Rakesh','Delhi')
```

**Updates in deletion vector table **

```python
%sql
update man_catlog.man_schema.delta_table2 set city = 'hyderabad'  where id = 1
```

**Optimize in delta tables**

**Deep Clone vs Shallow Clone**

**deep clone**
when you copy or create table using clone it will copy both metadata and data if you make changes in original table it wont effect in copied table

```python
%sql
create table text_ct.text_schema.deep_clone_table
deep clone man_catlog.man_schema.delta_table
```

```python
%sql
select * from text_ct.text_schema.deep_clone_table
```

```python
%sql
delete from man_catlog.man_schema.delta_table where id in (1,2,3,4,5,6)
```

```python
%sql
select * from man_catlog.man_schema.delta_table
```

```python
%sql
select * from text_ct.text_schema.deep_clone_table
```

```python
%sql
describe extended text_ct.text_schema.deep_clone_table
```

**Shallow Clone**
It will copy only metadata not the data and it will only support managed tables not external table

```python
%sql
create table text_ct.text_schema.shallow_clone 
shallow clone text_ct.text_schema.deep_clone_table
```

```python
%sql
describe extended text_ct.text_schema.shallow_clone
```
