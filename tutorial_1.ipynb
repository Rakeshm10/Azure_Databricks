### Scenario 1
**Managed Catalog-** 
**Managed Schema-**
**Managed Table**

**Managed Catalog**

```python
%sql
create catalog test_catalog
```

**Managed_Schema**

```python
%sql
create schema test_catalog.man_schema
```

```python
%sql DESCRIBE SCHEMA EXTENDED man_catlog.man_schema
```

**MAnaged_Table**

```python
%sql
create table test_catalog.man_schema.man_table (id int, name varchar(20)) using delta
```

```python
%sql
describe extended test_catalog.man_schema.man_table;
```

```python
%sql
insert into test_catalog.man_schema.man_table values(1,'rakesh');
```

```python
%sql
SELECT table_catalog, table_schema, table_name, table_type 
FROM system.information_schema.tables 
WHERE table_name = 'man_table' 
  AND table_schema = 'man_schema';

```

```python
%sql
select * from test_catalog.man_schema.man_table;
```


####Querying files using select

####Permanent Views

```python
%sql
create view man_catlog.man_schema.view1 
as
select * from delta.`abfss://mycontainer@storagedatabricks1005.dfs.core.windows.net/ext_table/ext_table1` where id < 4
```

```python
%sql
select * from man_catlog.man_schema.view1
```

####Temparary Views

**Temp view**

```python
%sql
create or replace temp view temp_view
as
select * from delta.`abfss://mycontainer@storagedatabricks1005.dfs.core.windows.net/ext_table/ext_table1` where id < 4
```

```python
%sql
select * from temp_view
```

####Volumes

**creating a directory using databricks notebook in adls**

```python
dbutils.fs.mkdirs('abfss://mycontainer@storagedatabricks1005.dfs.core.windows.net/volumes')
```

**creating a volume over directory and volume is equivalent to table in databricks**

```python
%sql
create external volume man_catlog.man_schema.ext_volume
location 'abfss://mycontainer@storagedatabricks1005.dfs.core.windows.net/volumes'
```

**copy file for volume**

```python
dbutils.fs.cp('abfss://mycontainer@storagedatabricks1005.dfs.core.windows.net/source/employees.csv','abfss://mycontainer@storagedatabricks1005.dfs.core.windows.net/volumes')
```

**Querying the volume**

```python
%sql
select * from csv.`/Volumes/man_catlog/man_schema/ext_volume/employees.csv`
```
