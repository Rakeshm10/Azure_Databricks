### Scenario 1
**Managed Catalog-** 
**Managed Schema-**
**Managed Table**

**Managed Catalog**

```python
%sql
create catalog man_catlog
```

**Managed_Schema**

```python
%sql
create schema man_catlog.man_schema
```

```python
%sql DESCRIBE SCHEMA EXTENDED man_catlog.man_schema
```

**MAnaged_Table**

```python
%sql
create table man_catlog.man_schema.man_table (id int, name varchar(20)) using delta
```

```python
%sql
describe extended man_catlog.man_schema.man_table;
```

```python
%sql
insert into man_catlog.man_schema.man_table values(1,'rakesh');
```

```python
%sql
SELECT table_catalog, table_schema, table_name, table_type 
FROM system.information_schema.tables 
WHERE table_name = 'man_table' 
  AND table_schema = 'man_schema';

```

```python
%sql
select * from man_catlog.man_schema.man_table;
```

### Scenario 2

```python
%sql
create schema ext_catalog.man_schema
```

```python
%sql
create table ext_catalog.man_schema.my_table(id int, name varchar(20))
```

```python
%sql
insert into ext_catalog.man_schema.my_table values(1,'rakesh');
```

### Scenario 3
**managed Catalog-** 
**managed Schema-**
**external Table**

```python
%sql
create table man_catlog.man_schema.ext_table(id int, name varchar(20))
 location 'abfss://mycontainer@storagedatabricks1005.dfs.core.windows.net/ext_table/ext_table1'
    
```

#### Drop Managed table

```python
%sql
drop table man_catlog.man_schema.man_table;
```

####Undrop managed table

```python
%sql
undrop table man_catlog.man_schema.man_table;
```
